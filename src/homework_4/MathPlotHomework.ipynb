```python
import random
import datetime
# import prettytable                  #   
import matplotlib.pyplot as plt     #   

def BubbleSort(A):                  #  
    for i in range(len(A)):
        for j in range(len(A)-1-i):
            if A[j] > A[j+1]:
                a = A[j]
                A[j] = A[j+1]
                A[j+1] = a


def QuickSort(A, fst, lst):         #  
    if fst >= lst:
        return

    #i, j = fst, lst
    pivot = A[fst]
    # pivot = A[random.randint(fst, lst)]
    first_bigger = fst+1
    while first_bigger <= lst:
        if A[first_bigger] >= pivot:
            break
        first_bigger += 1
    i = first_bigger+1
    while i <= lst:
        if A[i] < pivot:
            A[i], A[first_bigger] = A[first_bigger], A[i]
            first_bigger += 1
        i += 1

    last_smaller = first_bigger-1
    A[fst], A[last_smaller] = A[last_smaller], A[fst]
    QuickSort(A, fst, last_smaller-1)
    QuickSort(A, first_bigger, lst)

# def insert_sort(A):
#     for i in range(len(A)):
#         num = A[i]
#         j = i
#     while j > 0:
#         if A[j - 1] > num:
#             A[j] = A[j - 1]
#             j -= 1
#         else:
#             break
#     A[j] = num

def insert_sort(A):
    for i in range(1, len(A)): 
        num = A[i] 
        j = i-1
        while j >= 0 and num < A[j]:  
            A[j+1] = A[j]  
            j -= 1
        A[j+1] = num

def shaker_sort(A):
    for i in range(int(len(A) - 1 / 2)):
        for j in range(i, len(A) - 1 - i):
            if A[j] > A[j + 1]:
                A[j] = A[j + 1]
        for j in range(len(A) - 2 - i, i + 1):
            if A[j] < A[j - 1]:
                A[j] = A[j - 1]

# table = prettytable.PrettyTable([" ", " ", " "])
x=[]
y1=[]
y2=[]
y3=[]
y4=[]

for N in range(1000,5001,1000):
    x.append(N)
    min=1
    max=N
    A=[]
    for i in range (N):
        A.append(int(round(random.random()*(max-min)+min)))

    #print(A)

    B = A.copy()
    C = A.copy()
    D = A.copy()
    # print(B)

    # BubbleSort(A)
    print("---")
    # print(A)



    # QuickSort(B, 0, len(B)-1)
    print("---")
    # print(B)

    t1 = datetime.datetime.now()
    BubbleSort(A)
    t2 = datetime.datetime.now()
    y1.append((t2-t1).total_seconds())
    print("    " +str(N)+"      "+str((t2-t1).total_seconds()) + "c")

    t3 = datetime.datetime.now()
    QuickSort(B, 0, len(B)-1)
    t4 = datetime.datetime.now()
    y2.append((t4 - t3).total_seconds())
    print("   " +str(N)+"      "+str((t4-t3).total_seconds()) + "c")

    t5 = datetime.datetime.now()
    insert_sort(C)
    t6 = datetime.datetime.now()
    y3.append((t6 - t5).total_seconds())
    print("Insert   " +str(N)+"      "+str((t6-t5).total_seconds()) + "c")

    t7 = datetime.datetime.now()
    shaker_sort(D)
    t8 = datetime.datetime.now()
    y4.append((t8 - t7).total_seconds())
    print("Shaker   " +str(N)+"      "+str((t8-t7).total_seconds()) + "c")


    # table.add_row([str(N), str((t2-t1).total_seconds()), str((t4-t3).total_seconds())])
# print(table)

plt.title("Sorting algorithms")
plt.xlabel("elements")
plt.ylabel("time")

plt.plot(x, y1, "C0")
plt.plot(x, y2, "C1")
plt.plot(x, y3, "C2")
plt.plot(x, y4, "C3")
plt.show()

plt.title("Bubble sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.plot(x, y1, ':')
plt.show()

plt.title("Quick sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.plot(x, y2, ':')
plt.show()

plt.title("Insert sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.plot(x, y3, ':')
plt.show()

plt.title("Shaker sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.plot(x, y4, ':')
plt.show()

x = ["1000", "2000", "3000", "4000", "5000"]

plt.title("Bubble sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.bar(x, y1)
plt.show()

plt.title("Quick sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.bar(x, y2)
plt.show()

plt.title("Insert sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.bar(x, y3)
plt.show()

plt.title("Shaker sort")
plt.xlabel("elements")
plt.ylabel("time")
plt.bar(x, y4)
plt.show()

```
